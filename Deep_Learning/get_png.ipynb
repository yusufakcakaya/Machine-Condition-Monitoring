```python
#Import necessary  libraries
import librosa
from sklearn.preprocessing import scale
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import os
```

```python
def to_png(files,directory,name_file):    

    
    data, sr = librosa.load(files)
    data = scale(data)
           
    melspec = librosa.feature.melspectrogram(y=data, sr=sr, n_mels=128)
           
    log_melspec = librosa.power_to_db(melspec, ref=np.max)  
    librosa.display.specshow(log_melspec, sr=sr)
            
        
      
    if not os.path.exists(directory):
        os.makedirs(directory)


    plt.savefig(directory + '/' + name_file + '.png')            
    
    return 0
```

```python
path = 'AudiOutofDataset/normal.m4a'
to_png(path,'new_test','audio1')
```

```python
path = 'AudiOutofDataset/tiktik.m4a'
to_png(path,'new_test','audio2')
```
